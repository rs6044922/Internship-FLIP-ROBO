**Q1: Write a python program to scrape data for “Data Analyst” Job position in
“Bangalore” location. You have to scrape the job-title, job-location, company_name,
experience_required. You have to scrape first 10 jobs data.**

```python
#Calling driver to load the website
driver=webdriver.Chrome('chromedriver.exe')
url='https://www.naukri.com/'
driver.get(url)

#Calling driver to load the website
search_job=driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data Analyst')
search_loc=driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys('Bangalore')
button_clk=driver.find_element_by_class_name('search-btn')
button_clk.click()
```

```python
#Searching the target webpage element
titles_job=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
job_title=[]
for i in titles_job:
    job_title.append(i.text)
job_title=job_title[0:10]

title_loc=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
job_loc=[]
for i in title_loc:
    job_loc.append(i.text)
job_loc=job_loc[0:10]

title_company=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
com_name=[]
for i in title_company:
    com_name.append(i.text)
com_name=com_name[0:10]

title_exp=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
can_exp=[]
for i in title_exp:
    can_exp.append(i.text)
can_exp=can_exp[0:10]
```

```python
#Creating Dataframe
Data_analyst=pd.DataFrame({})
Data_analyst['Job-Title']=job_title
Data_analyst['Job-Location']=job_loc
Data_analyst['Company_Name']=com_name
Data_analyst['Experience_Required']=can_exp
Data_analyst
```

```python
#Searching the target webpage targeted element
title_company=driver.find_elements_by_xpath("//a[@class='css-f3vw95 e1aj7ssy3']")
com_name=[]
for i in title_company:
    com_name.append(i.text)
com_name=com_name[0:10]

title_nosalary=driver.find_elements_by_xpath("//div[@class='col-12 col-lg-auto']")
salary_no=[]
for i in title_nosalary:
    salary_no.append(i.text)
salary_no=salary_no[0:10]

title_avgsalary=driver.find_elements_by_xpath("//div[@class='col-12 col-lg-4 px-lg-0 d-flex align-items-baseline']")
salary_avg=[]
for i in title_avgsalary:
    salary_avg.append(i.text.replace('\n /yr','').strip())
salary_avg=salary_avg[0:10]

title_minsal=driver.find_elements_by_xpath("//div[@class='d-flex mt-xxsm css-79elbk epuxyqn0']")
salary_min=[]
for i in title_minsal:
    salary_min.append(i.text.split('\n'))
salary_min=salary_min[0:10]
salary_min

max=[]
min=[]
for i in range(len(salary_min)):
    for j in range(len(salary_min[i])):
        if j == 1:
            max.append(salary_min[i][j])
        else:
            min.append(salary_min[i][j])
```

```python
#Creating Dataframe
Data_Scientist=pd.DataFrame({})
Data_Scientist['Company_Name']=com_name
Data_Scientist['Number of salaries']=salary_no
Data_Scientist['Average salary']=salary_avg
Data_Scientist['Minsalary']=min
Data_Scientist['Maxsalary']=max

Data_Scientist
```

 **Q7: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to
go the link: https://www.flipkart.com/apple-iphone-11-black-64-gb-includesearpods-poweradapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace.**

```python
#Calling driver to load the website
driver=webdriver.Chrome('chromedriver.exe')
url='https://www.flipkart.com/apple-iphone-11-black-64-gb-includesearpods-poweradapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace.'
driver.get(url)
```

```python
driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']/span").click()
```

```python
#Searching the target webpage targeted element
import time
Rating = []
Review_summary = []
Full_review = []

#Scraping 100 Reviews
while True:
    time.sleep(3)
    
    for i in driver.find_elements_by_xpath("//div[@class = 'col _2wzgFH K0kLPL']/div[@class= 'row']/div"):
        Rating.append(i.text)
        if len(Rating) == 200:
            break 
            
    for i in driver.find_elements_by_xpath("//div[@class= 'row']//p[@class = '_2-N8zT']"):
        Review_summary.append(i.text)
        if len(Review_summary) == 100:
            break 
            
    for i in driver.find_elements_by_xpath("//div[@class= 't-ZTKy']//div[@class]"):
        Full_review.append(i.text.replace('\n',''))
        if len(Full_review) == 100:
            break 
        
    if len(Full_review) == 100:
        break    
                
    for i in driver.find_elements_by_xpath("//nav[@class='yFHi8N']//a[@class='_1LKTO3']"):
        next_page_1=[i.get_attribute("href")]
    
    time.sleep(3)
    
    for n in next_page_1:
        driver.get(n)

driver.close()
```

```python
Ratings = []
for i in range(0,len(Rating)):
    if i == 0 or  i/2 == i//2:
        Ratings.append(Rating[i])
```

```python
#Creating Dataframe
flipkart=pd.DataFrame({})
flipkart['Rating']=Ratings
flipkart['Review_summary']=Review_summary
flipkart['Full_review']=Full_review

pd.set_option("display.max_colwidth",None)
flipkart
```

```python
#Creating lists for attributes
Brand_1 = []
Product_1= []
Price_1 = []
Discount_1 =[]

#Scraping 100 sneakers data
while True:
    
    time.sleep(3)
        
    for i in driver.find_elements_by_xpath("//div[@class= '_2B099V']//div[@class = '_2WkVRV']"):
        Brand_1.append(i.text)
        if len(Brand_1) == 100:
            break
    
    for i in driver.find_elements_by_xpath("//div[@class= '_2B099V']//a"):
        Product_1.append(i.get_attribute('title'))
        if len(Product_1) == 200:
            break
    
    for i in driver.find_elements_by_xpath("//div[@class= '_25b18c']//div[@class = '_30jeq3']"):
        Price_1.append(i.text) 
        if len(Price_1) == 100:
            break    
    for i in driver.find_elements_by_xpath("//div[@class= '_3Ay6Sb']/span"):
        Discount_1.append(i.text)
        if len(Discount_1) == 100:
            break  
            
    if len(Discount_1) == 100:
        break
        
    for j in driver.find_elements_by_xpath("//nav[@class ='yFHi8N']/a[@class = '_1LKTO3']"):
        next_page_2=[j.get_attribute("href")]
    
    time.sleep(3)
        
    for n in next_page_2:
        driver.get(n) 
    
driver.close()
```

```python
#Shortening Product Description by removing unwanted datas
Product_Desc = []
for i in range(0,len(Product_1)):
    if i == 0 or  i/2 == i//2:
        Product_Desc.append(Product_1[i])
```

```python
flipkart=pd.DataFrame({})
flipkart['Brand']=Brand_1
flipkart['Product Description']=Product_Desc
flipkart['Price']=Price_1
flipkart['Discount']=Discount_1

flipkart
```
